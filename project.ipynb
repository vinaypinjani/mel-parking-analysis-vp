```python
# Importing relavant libraries
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# Creating a data frame
# Working with only 10% of the data due to file size
sensorData = pd.read_csv('On-street_Car_Parking_Sensor_Data_-_2020__Jan_-_May_.csv').sample(frac=0.1, random_state=123)
```

```python
# Identifying the features
print(sensorData.shape)
print(sensorData.dtypes)
sensorData
```

```python
# Identifying numeric columns
df_numeric = sensorData.select_dtypes(include=np.number)
numeric_cols = df_numeric.columns.values
print(numeric_cols)
```

```python
# Identifying non numeric columns
df_non_numeric = sensorData.select_dtypes(exclude=np.number)
non_numeric_cols = df_non_numeric.columns.values
print(non_numeric_cols)
```

```python
# Converting time stamps into datetime stamps
sensorData['ArrivalTime'] = pd.to_datetime(sensorData['ArrivalTime'])
sensorData['DepartureTime'] = pd.to_datetime(sensorData['DepartureTime'])
```

```python
# Dropping non-reqired columns
sensorData = sensorData.drop(columns = ['StreetMarker', 'SignPlateID', 'BetweenStreet1ID', 'BetweenStreet1', 'BetweenStreet2ID', 'BetweenStreet2', 'SideOfStreet'])
```

```python
# Updated data types
sensorData.info()
```

Questions
1. Busiest street in rush hours?
2. Average parking time for major streets?
3. Areas where most violations occur?
4. Streets with highest average parking time?
5. Which day of the week is busiest?

```python
# Finding streets with most most sensor triggers 
sensorData.groupby(['StreetId', 'StreetName']).size().sort_values(ascending=False)
```

## Londsale Street is the most parked street in 2020

```python
# Analysing morning rush hours
dateData = sensorData.set_index("ArrivalTime")
dateData1 = dateData.between_time('06:00:00', '10:00:00')
dateData1.groupby(['StreetId', 'StreetName']).size().sort_values(ascending=False)
```

```python
# Analysing afternoon rush hours
dateData2 = dateData.between_time('15:00:00', '17:00:00')
dateData2.groupby(['StreetId', 'StreetName']).size().sort_values(ascending=False)
```

## Londsdale street has the most cars parked on rush hours.

```python
# Dataframe for 3 main streets
ids = [894, 528, 839]
mainStreets = sensorData[sensorData.StreetId.isin(ids)]
mainStreets
```

```python
# Average parking duration for main streets
mainStreets['DurationMinutes'].mean()
```

# Average time of parking on major streets(Londsdale Street, Kings Street, Collins Street) is 30 minutes.

```python
# Generating total violations by area
areaData = sensorData.groupby("AreaName")['InViolation'].sum()
areaData.sort_values(ascending=False)
```

## Docklands has significantly more violations than any other area.

```python
# Highest average parking time by streets
avgTime =  sensorData.groupby(['StreetId', 'StreetName'])['DurationMinutes'].mean()
avgTime.sort_values(ascending=False)
```

 ## LITTLE DRYBURGH STREET SOUTH has highest average parking time that is around 5 hours

```python
# Sorting data by days and counting them
dayData = sensorData['ArrivalTime'].dt.dayofweek
dayData.groupby(dayData.values).size().sort_values(ascending=False)
```

## Friday has the most number of sensor triggers
